```python
import random
class niño:
  def __init__(self,nombre,sexo,papa,mama):
    self.nombre = nombre
    self.sexo = sexo
    self.papa = papa
    self.mama = mama
    self.id = self.id()

  def id(self):
    num = 0
    while len(str(num))<2:
       num = int(random.random()*100)
       print(num)
    return num
  
  def prt(self):
    rt = self.nombre+' '+self.sexo+' '+self.papa+' '+self.mama
    return rt  

```

```python
class pais():
  def __init__(self,nombre):
    self.nombre = nombre
    self.ciud = []

  def new_p(self,nombre,sexo,papa,mama):
    band = True
    while band:
      ni = niño(nombre,sexo,papa,mama)
      band = self.verif(ni)
    self.ciud.append(ni)  

  def verif(self, ni_v):
    for cd in self.ciud:
      if ni_v.id == cd.id:
        return True
    return False


clb = pais('Colombia')
```
