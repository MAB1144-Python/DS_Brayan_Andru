```python
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons, make_circles
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt 
```

```python
make_circles()
```

```python
X,y = make_blobs(n_samples= 500, n_features=3, centers= 2, cluster_std = 3, random_state=10)
```

```python
fig,axs = plt.subplots(11,3, figsize = (12,40))#[1,2,3,4]

axs[0][0].scatter(X[:,0],X[:,1],c = y)
axs[0][1].scatter(X[:,1],X[:,2],c = y)
axs[0][2].scatter(X[:,0],X[:,2],c = y)
for i in range(1,11):
  dbscan_cluster = DBSCAN(eps = 2, min_samples = 5)
  dbs = dbscan_cluster.fit_predict(X)
  dbscan_cluster.components_

  kmeans_cluster = KMeans(n_clusters = 2, max_iter= i,
                          init='random')
  clusters = kmeans_cluster.fit_predict(X)
  center = kmeans_cluster.cluster_centers_
  clusters

  axs[i][0].scatter(X[:,0],X[:,1],c = clusters)
  axs[i][0].scatter(center[:,0],center[:,1])
  axs[i][1].scatter(X[:,1],X[:,2],c = clusters)
  axs[i][1].scatter(center[:,1],center[:,2])
  axs[i][2].scatter(X[:,0],X[:,2],c = clusters)
  axs[i][2].scatter(center[:,0],center[:,2])
```

```python
aglomer_cluster = AgglomerativeClustering(n_clusters=2)
aglo = aglomer_cluster.fit_predict(X)
aglo
```

```python
fig,axs = plt.subplots(1,3, figsize = (12,4))#[1,2,3,4]

axs[0].scatter(X[:,0],X[:,1],c = aglo)
axs[1].scatter(X[:,1],X[:,2],c = aglo)
axs[2].scatter(X[:,0],X[:,2],c = aglo)
```

```python
Xm,ym = make_moons(n_samples= 100, noise= 0.08, random_state = 10)
fig,axs = plt.subplots(1,4, figsize = (16,4))#[1,2,3,4]

axs[0].scatter(Xm[:,0],Xm[:,1],c = ym)
km = KMeans(n_clusters=2)
km_t = km.fit_predict(Xm)
axs[1].scatter(Xm[:,0],Xm[:,1],c = km_t)

ag = AgglomerativeClustering(n_clusters=2)
ag_t = ag.fit_predict(Xm)
axs[2].scatter(Xm[:,0],Xm[:,1],c = ag_t)

db = DBSCAN(eps = 0.30, min_samples= 3)
db_t = db.fit_predict(Xm)
axs[3].scatter(Xm[:,0],Xm[:,1],c = db_t)
plt.show()
```
